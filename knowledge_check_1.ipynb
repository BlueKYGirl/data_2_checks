First, let's install necessary libraries and modules to work with an API to bring in our data.

```python
import pandas as pd
import requests
```

Now let's bring in data from a national COVID-19 tracking API. We will use the state data endpoint.

```python
r = requests.get('https://api.covidtracking.com/v1/states/daily.json')
data = r.json()
df = pd.DataFrame(data)
df.head()
```

Great! Let's get some information and descriptive statistics about our data. First, let's cast a wide net and describe the entire data frame so we can see the counts, means, maxes, and mins of all of our numeric data.

```python
df.describe()
```

Next, let's use the info() function to find column dtypes and count non-null values. This data frame has a large number of columns, so let's use the verbose parameter to show all columns.

```python
df.info(verbose=True)
```

For our required query, we want to see Kentucky's COVID-19 historical data only. 

```python
df[df['state'] == 'KY']
```

Now we will select and print the second and third columns of the data frame.

```python
df.iloc[:, 1:3]
```

Finally, let's select and print the first 4 rows of the data frame.

```python
df.iloc[0:4]
```
